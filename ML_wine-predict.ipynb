```python
# Carregando dataset
import pandas as pd
arquivo = pd.read_csv('D:/Python/Kaggle/wine_dataset.csv')
```

```python
arquivo.head()
```

```python
# Para trabalhar com a classificação e fazer os cálculos depois, precisamos transformar a variável style
# Style está em texto, precisamos passar para número
```

```python
# Transformar 'red' em 0
arquivo['style'] = arquivo['style'].replace('red', 0)
```

```python
# Transformar 'white' em 1
arquivo['style'] = arquivo['style'].replace('white', 1)
```

```python
# Visualizar transformação
arquivo
```

```python
# Separar as variáveis entre preditoras e variável alvo (aquila que queremos prever)
# y recebe a coluna "style" e x recebe todas as outras
y = arquivo['style']
x = arquivo.drop('style', axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Criando os conjuntos de dados de treino e teste:
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.33,
                                                    random_state=0,
                                                    stratify=y)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
# Criação do Modelo:
modelo = ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_train, y_train)

# Imprimindo resultados:
resultado = modelo.score(x_test, y_test)
print("Acurácia:", resultado)
```
